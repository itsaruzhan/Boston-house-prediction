```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston

boston = load_boston()
boston
```

```python
df = pd.DataFrame(data=boston.data, columns=boston.feature_names)
df["price"] = boston.target
```

```python
df
```

```python
df.describe()
```

```python
# RAD and TAX have a 0.91 correlation. When training the model, we should not use both of these features together, or decrease this high value
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# We see from these histplots that TAX and RAD have too high values(for RAD there is no data between 10 and 23, for TAX most of the data lies between 0 and 450) 
df.hist(figsize=(15, 10), bins=20)
plt.tight_layout();
```

```python
sns.histplot(data=df, x="RAD", bins=20)
plt.show()
```

```python
data = df[df['RAD']<20]
```

```python
# since RAT and TAX was high correlated features, we deleted high outliers of TAX, while drop rows that contain too high RATE 
sns.histplot(data=data, x="TAX", bins=20)
plt.show()
```

```python
# but I lose 26% of the data
(1-len(data)/len(df))*100
```

```python
corr = df.corr()
corr.style.background_gradient()
```

```python
# from the scatter plot of LSTAT and PRICE we see that the graph is similar to a logorithmic graph rather than a linear one
sns.scatterplot(x="LSTAT", y="price", data=data)
```

```python
data['LOG_LSTAT']= np.log(data['LSTAT'])
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(16,6))

axes[0].plot(data['LSTAT'],data['price'],'o',alpha=0.5)
axes[0].set_ylabel("price")
axes[0].set_title("LSTAT")

axes[1].plot(data['LOG_LSTAT'],data['price'],'o',alpha=0.5)
axes[1].set_ylabel("price")
axes[1].set_title("LOG_LSTAT")


plt.tight_layout();
```

```python
#Log transformation helps reducing skewness when you have skewed data.
fig, axs = plt.subplots(nrows=1, ncols=2)
lstat = data[['LSTAT']]
sq_lstat = data[['LOG_LSTAT']]
axs[0].hist(lstat, edgecolor='black')
axs[1].hist(sq_lstat,  edgecolor='black')

axs[0].set_title('Original Data')
axs[1].set_title('Log Transformed Data')
```

```python
data = data.drop(['LSTAT'], axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X=data.drop(['price'],axis=1)
y=data['price']
x_train,x_test,y_train,y_test= train_test_split(X,y, random_state=42,test_size=0.25)
linear = LinearRegression()
linear.fit(x_train, y_train)
linear.score(x_test, y_test)
```
